```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exploration 

```python
df = pd.read_csv(r"CarPrice_Assignment.csv")
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.drop(['car_ID'], axis =1, inplace = True) #the values are all unique
```

```python
#Isolating only company names
df['CarName']= df['CarName'].apply(lambda x:x.split(' ', 1)[0])
df.rename(columns = {'CarName':'company'}, inplace = True)
```

```python
df.head()
```

```python
df['company'].unique() #Found errors in names
```

```python
#Fixing names
df.company =df.company.str.lower()
def replace_name(a,b):
    df.company.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

df.company.unique()
```

```python
df.company.nunique()
```

```python
#Checking for Null values
null = df.isnull().sum()
print(null)
```

```python
df.drop(['compressionratio','carwidth','curbweight','wheelbase','citympg'], axis =1, inplace = True)
```

```python
df.head()
```

```python
df_dummy = df.select_dtypes(exclude=[np.number])
df_dummy
```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df_dummy.columns :
    df[col] = le.fit_transform(df[col])
df
```

# Training 


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X= df.iloc[:,:-1].values
y=df.iloc[:,-1].values

print(X)
```

```python
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.2,random_state=0)
model =LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_val)
```

```python
from sklearn.metrics import r2_score
r2_score(y_val,y_predicted)
```
